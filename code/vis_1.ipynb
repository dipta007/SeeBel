```python
def label2color(label, color_array):
    temp = label.clone()
    label1 = label.clone()
    label2 = label.clone()
    label3 = label.clone()
    #print (label.dtype, temp.dtype, label1.dtype, label2.dtype, label3.dtype)
    for k, v in color_array.items():
        label1[temp == k] = v[0]
        label2[temp == k] = v[1]
        label3[temp == k] = v[2]
    #print (label1.shape, label2.shape, label3.shape)
    label1 = rearrange(label1, 'd0 d1 d2 -> d0 1 d1 d2')
    label2 = rearrange(label2, 'd0 d1 d2 -> d0 1 d1 d2')
    label3 = rearrange(label3, 'd0 d1 d2 -> d0 1 d1 d2')
    #print (label1.shape, label2.shape, label3.shape)
    ret = torch.stack((label1, label2, label3), 1)
    ret = rearrange(ret, 'd0 d1 d2 d3 d4-> d0 (d1 d2) d3 d4')
    return ret
```

```python
with open(f'{DATA_DIR}/class_masks.pickle', 'rb') as fp:
    class_masks = pickle.load(fp)

with open(f'{DATA_DIR}/iou_masks.pickle', 'rb') as fp:
    iou_masks = pickle.load(fp)

with open(f'{DATA_DIR}/size_masks.pickle', 'rb') as fp:
    size_masks = pickle.load(fp)
```

```python
# Prepare all data at once
# For ALL
size = 1.1
legend_list = []
color_dict = {}
for cls, col in color_mapping_after_conversion.items():
    if (0<=cls and cls<NUM_CLASSES):
        temp = []
        for i in col:
            temp.append(i/255)
        temp.append(1.0)
        color_dict[cls] = temp
        legend_list.append(mpatches.Patch(color=color_dict[cls], label=id2class_after_conversion[cls]))

color_mask = [color_dict[cls] for cls in class_masks]
label_mask = [id2class_after_conversion[cls] for cls in class_masks]
colors = [[0.5, 1, 0, 1], [0.4, 0.5, 1, 1], [1, 0, 0, 1]]

sizes = [size*1.0 for val in class_masks]

# For selected objects
label_mask = [id2class_after_conversion[cls] for cls in class_masks]
labels = list(set(label_mask))

mp_iou_list = {}
mp_size_list = {}
mp_color_list = {}
mp_sizes = {}

for current_object in labels:
    new_iou_list = []
    new_size_list = []
    new_color_list = []
    new_sizes = []
    current_class = class2id_after_conversion[current_object]
    for ind, cls in enumerate(class_masks):
        if cls == current_class:
            new_iou_list.append(iou_masks[ind])
            new_size_list.append(size_masks[ind])
            new_color_list.append(color_mask[ind])
            new_sizes.append(1.0)
    
    mp_iou_list[current_class] = new_iou_list
    mp_size_list[current_class] = new_size_list
    mp_color_list[current_class] = new_color_list
    mp_sizes[current_class] = new_sizes
```

```python
def plot_stat(current_object="All"):
    global sizes
    fig, ax = plt.subplots(1, 2, figsize=(10,5))
    fig.dpi = 150

    # set figure texts
    caption = "Correlation between mask size (pixel area) and correctness of the AI model in predicting masks (IoU) for all masks in the Cityscapes Dataset's trainset."
    title = fig.suptitle("\n".join(wrap(caption, 120)), fontsize=10)
    ax[0].set_xlabel('Object Size')
    ax[0].set_ylabel('Object IoU')
    ax[1].set_xlabel('Object Size')
    ax[1].set_ylabel('Object IoU')
    ax[0].set_title('ALL')
    ax[1].set_title(current_object)


    # set axis limit
    xmn, xmx = min(size_masks), max(size_masks)
    ymn, ymx = min(iou_masks), max(iou_masks)
    ax[0].set_xlim([xmn, xmx])
    ax[1].set_xlim([xmn, xmx])
    ax[0].set_ylim([ymn, ymx])
    ax[1].set_ylim([ymn, ymx])
    
    # All Objects plotting
    ax[0].scatter(size_masks, iou_masks, s=sizes, c=color_mask)
    plt.legend(handles=legend_list, bbox_to_anchor=(1.4, 1.02), loc=1, 
                facecolor='white', framealpha=0.5)
    
    
    # Selected Object plotting
    current_class = class2id_after_conversion[current_object]
    new_size_list = mp_size_list[current_class]
    new_iou_list = mp_iou_list[current_class]
    new_sizes = mp_sizes[current_class]
    new_color_list = mp_color_list[current_class]
    
    ax[1].scatter(new_size_list, new_iou_list, s=new_sizes, c=new_color_list)

    plt.tight_layout()
    plt.show()

dropdown = widgets.Dropdown(options=labels, 
                            value='road', 
                            description='Object:')

interactive(plot_stat, current_object=dropdown)
```
